# 인구 피라미드

# 남자 데이터

```python
import pandas as pd
file_name = '201507_201507_연령별인구현황_월간.xlsx' # 다른 연도 데이터를 보고 싶을 때 유용함
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 1,111,111 -> 1111111 (정수형)
```

```python
df_m.iloc[0]
```

# 여자 데이터

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)  # 0~4세라는 데이터가 앞에도 있고 뒤에도 있기에 중복 방지
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w.head(3)
```

# 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명, 남자 음수로 표현
plt.barh(df_w.columns, df_w.iloc[0] // 1000) # 여자 양수로 표현
plt.title('2015 대한민국 인구 피라미드')
plt.savefig('2015 대한민국 인구 피라미드', dpi=100)
plt.show()
```
